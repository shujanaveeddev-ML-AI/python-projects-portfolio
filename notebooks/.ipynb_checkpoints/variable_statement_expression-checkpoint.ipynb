# Introduction to Python: Variables, Statements, and Expressions


## Comments


Comments are non executable statements i.e ignored by interpreter. They are used to take notes in the code. They start with a # sign which tell compiler to ignore evrythig that comes after it.


```python
print("Hello World!") # print function is used to output statement to the terminal.
```

## Expression

Expression is a combination of literals variable and function call that get evaluated by python.
Expression -> Value -> Type

```python
print(100) # print(100) -> 100 -> int
print("Hello World") # print("Hello World") -> 100 -> int
```

## Operators

Operators are tokens that are used to specify computation like add, sub etc.

### Arithmetic Operators

#### Addition (+)

Add the value of the two expression

```python
print(100+200) # Value: 300
```

#### Subtraction (-)

Subtract the value of operand on the right from the operand on the left

```python
print(100-22)
```

#### Multiplication (*)

Multiply the value of two operands

```python
print(10*3)
```

#### Division (/)

Divide the value of operand on the right from the operand on the left

```python
print(10/3)
```

Whenever we do normal division in python we get a float back

#### Floor Division (//)

Divide the value of two expressions and truncate the decimal

```python
print(10//3)
```

If one of the operand is float. It produces a float even though it truncates all the decimal places.

#### Exponentiation (***)

Left operand raised to the power Right operand.

```python
print(4**2) # 4 raised to the power 2
```

#### Modulo (%)

Work on integers. Yield remainder as left operand get divided by right.

```python
print(7%3) # Value: 1
```

### Compound Assignment Operators

Combine assignment operator with arithmetic operator. Provide a short syntax to update the value of variable using old variables

```python
x = 3
print(x)
```

#### Add and Assign (+=)

Add value of operand on the right to operand on left and assign it to the left operand.

```python
x+=3
print(x)
```

#### Subtract and Assign (-=)

Subtract value of operand on the right from operand on left and assign it to the left operand.

```python
x-=2
print(x)
```

#### Multiply and Assign (*=)

Multiply value of operand on the right to operand on left and assign it to the left operand.

```python
x*=3
print(x)
```

#### Divide and Assign (/=)

Divide value of operand on the right from operand on left and assign it to the left operand

```python
x/=3
print(x)
```

#### Modulus and Assign (/=)

Find the remainder when the operand on the left is divided by the operand on the right and assign it to the left operand.

```python
x %= 8
print(x)
```

#### Exponentiate and Assign (**=)

Raise the left operand to the power of the right operand and assign the result to the left operand.

```python
x **= 8
print(x)
```

#### Floor Division and Assign (**=)

Perform floor division of the left operand by the right operand and assign the result (integer part) to the left operand.

```python
y = 28
y//=8
print(y)
```

### Order of Operations

If an expression contain multiple operation then it helps us to decide which operation is performed first. The order of operation in python is the same as in maths.

#### Order in Python

() > ** > unary (+ - !) > * / % > + - > << >> > < <= > >= > == != > & > ^ > | > && > || > ?: > (and compound assignments)

## Variable

 Variable are useful as they allow you to store the value computed by an exression. They allow you to assign human readable name to variable. It allow you to used the value again  

### Setting variable

```python
message = "What's up, Shuja?" 
n = 17
pi = 3.14
```

### Referencing Variable

```python
print(message)
print(n)
print(pi)
```

### Reassigning a Variable

In python once a variable has been initialized you can reassign value to it.

```python
# Set the value of x to 5 and then print it
x= 5
print(x) # Result: 5

# Reassign the value of x to 10 and then print it.
x = 10
print(x)
```

### Rule for Naming a Variable


1. Start with a letter.
2. Contains letter or numbers.
3. Can contain undercore ( _ ).
4. Cannot be a python special keyword.
5. Python is case sensitive i.e x != X.

## Updating Variable

You can update the value of variable once it has been created. You can also use the old value of variable to get the new value of variable.

```python
x = 6           # Initialize x
print(x)
x +=3          # Increment x by 3, same as x= x+3
print(x)
x -=1          # Decrement x by 1, same as x= x-3
print(x)
```

When we make a function call the following steps get execution by Python behind the scene:

#### Converting Seconds to Hours, Minutes and Seconds

```python
str_second = input("Please enter the numbers of second you wish to convert:")
total_seconds = int(str_second) # Convert str_second of type str to int 

hours = total_seconds // 3600                # Calculate the no of hours
sec_remaining = total_seconds % 3600         # Find the remainder seconds when seconds get converted to hours
minutes = total_seconds // 60                #  Calculate the no of minutes using remainder seconds
sec_finally_remaining = total_seconds % 60   #  Find the final remainder seconds when seconds get converted to minutes

print("Hrs=", hours, "mins=",minutes,"secs= ",sec_finally_remaining)
```
