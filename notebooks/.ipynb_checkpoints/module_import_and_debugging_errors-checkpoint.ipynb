# Importing Modules and Understanding Python Errors

## Importing Module

Python comes with a lot of package that contain many function that provides us with a lot of powerful feature. Some of these are built in python and some you need to install using pip module. The import statement is used to import these module that coontain those functions.

```python
import random

prob = random.random()  # Return a random float between 0 and 1 excluding 1
print(prob)
```

random.random() mean a function named random inside random module.

```python
diceThrow = random.randrange(1,7)    # return int from 1 to 6
print(diceThrow) 
```

It is a import to note that **random number generator** are based on **detreministic algorithm** - repeatable and predictable. So they are called **pseudo random generator**. They start with a **seed value**. Each time you ask for another number you will get one based on currents **seed attribute**.

### Importing Functions Rather Than Entire Module

We can use the syntax **from module import functions** to import functions individuallay than the entire module. 

```python
from random import randrange
randnum = randrange(10)            # return int from 0 to 9
print(randnum)
```

**Note:** If you are importing a single function than you should **call the function directly** than using the **module.function** syntax.

The run time error occur when it is able to parse through the code but an error occur on the **run time** due to an **illegal operation**.

```python
print(3/0)
```

#### Description

You are performing an illegal operation of dividing by zero which is unsolveable.

### Semantic Error

This error occur when the interpreter is able to parse throught the code and it is executed succesfully but the expected output is not produced.

```python
num1 = input('Enter a number:')
num2 = input('Enter another number:')
sum = num1 + num2

print('The sum of', num1, 'and', num2, 'is', sum)
```

### Value Error

This error when your function works for a range of values but you input a value outside that range

```python
current_time_str = input("What is the current time (in hours 0-23)?")
current_time_int = int(current_time_str)

wait_time_str = input("How many hours do you want to wait")
wait_time_int = int(wait_time_str)

final_time_int = current_time_int + wait_time_int
print(final_time_int)
```

#### Description

You have entered a value outside the range of function so now it will produce inaccurate output.

#### Fix

Check the constraint of values of a function can accept.
